```python
import lifesim
```

```python
# create bus
bus = lifesim.Bus()

# setting the options
bus.data.options.set_scenario('baseline')

# set options manually
bus.data.options.set_manual(output_path='/Users/fdannert/Documents/projects/LIFEsim/working/requests/eckhart/output/')
bus.data.options.set_manual(output_filename='test_run')

bus.data.catalog_from_ppop(input_path='/Users/fdannert/Documents/projects/LIFEsim/working/requests/jonah_hansen/scan_parameters/inputs/small_catalog_lifesim_tutorial.txt')
```

```python
cat = bus.data.catalog
cat['snr_1h'] = 0.
```

```python
cat.head()
```

```python
transm = lifesim.TransmissionMap(name='transm')
bus.add_module(transm)

opt = lifesim.Optimizer(name='opt')
bus.add_module(opt)
ahgs = lifesim.AhgsModule(name='ahgs')
bus.add_module(ahgs)

bus.connect(('transm', 'opt'))
bus.connect(('inst', 'opt'))
bus.connect(('opt', 'ahgs'))
```
