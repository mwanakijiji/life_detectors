```python
# print a random subsample based on a condition
N_sample = 10

# condition
rows_of_interest = df.index[df['Stype'] == 'G']

# Randomly sample 10 indices from rows_with_A (without replacement)
if len(rows_of_interest) >= N_sample:
    random_indices = np.random.choice(rows_of_interest.values, size=N_sample, replace=False)
else:
    random_indices = rows_of_interest.values  # If there are fewer than 10, just use all

print("10 random row indices with Stype ", random_indices)

# write out the file with the same two rows as the file with name file_name, and use the same formatting (delim_whitespace=True, etc.)
# Write out a file with the same first two lines as the source file, followed by the N_sample randomly selected rows for Stype == 'A'

# The name of the output file
out_file = "/Users/eckhartspalding/Documents/git.repos/life_detectors/felix_yields/input/junk_sample_Stype_.txt"

# Read in the first two lines from the original file (header rows)
with open(file_name, 'r') as f:
    header1 = f.readline()
    header2 = f.readline()

# Subset the DataFrame for the sampled indices and format as space-delimited text
subset_df = df.loc[random_indices]

# Write to the output file
with open(out_file, 'w') as f:
    f.write(header1)
    f.write(header2)
    # Save the selected rows, space-delimited, no index, float format as in input
    subset_df.to_csv(f, sep=' ', header=False, index=False, float_format="%.6g")

print(f"Wrote {len(subset_df)} rows with Stype == 'A' to {out_file}")



#for idx in rows_with_A.values:
#    print(idx)
```
