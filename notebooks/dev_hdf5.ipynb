```python
df = pd.read_hdf(file_name, key="catalog")   # or "/catalog"
print(df.head())
```

```python
df.keys()
```

```python
# save this as a text file with column names modified to enable drop-in functionality

# stay the same: fp, z                                              
df_renamed = df.rename(columns={'radius_p': 'Rp', 
                                'nuniverse': 'Nuniverse',
                                'nstar':'Nstar',
                                'stype':'Stype',
                                'mass_s':'Ms',
                                'mass_p':'Mp',
                                'ra': 'RA', 
                                'dec': 'DEC', 
                                'name': 'name_s', 
                                'lat': 'bGal', 
                                'lon': 'lGal', 
                                'temp_s': 'Ts',
                                'distance_s': 'Ds', 
                                'radius_s': 'Rs', 
                                'temp_p': 'Tp', 
                                'flux_p': 'Fp', 
                                 'ecc_p': 'ep', 
                                 'inc_p': 'ip', 
                                 'large_omega_p': 'Omegap', 
                                 'small_omega_p': 'omegap', 
                                 'theta_p': 'thetap', 
                                 'angsep': 'AngSep', 
                                 'maxangsep': 'maxAngSep', 
                                 'p_orb': 'Porb', 
                                 'albedo_bond': 'Abond', 
                                 'albedo_geom_vis': 'AgeomVIS', 
                                 'albedo_geom_mir': 'AgeomMIR', 
                                 'semimajor_p': 'ap', 
                                 'sep_p': 'rp'})
```

```python
df_renamed['Ts']
```

```python
# Write DataFrame to CSV with a comment as the first line
output_csv = "junk_catalog_renamed.csv"
comment_line = "# Catalog data reformatted for drop-in functionality\n"
with open(output_csv, "w") as f:
    f.write(comment_line)
    df_renamed.to_csv(f, index=False, sep=' ')
print(f"Saved CSV with comment to {output_csv}")
```
